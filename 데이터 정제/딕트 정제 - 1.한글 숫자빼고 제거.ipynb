```python
from collections import Counter

f = open("./dict/nia.txt", 'r')
lines = f.readlines()
f.close()

data = ""
for line in lines:
    data = data + line
print(len(set(data)),set(data))
```

```python
import re
## 한글과 띄어쓰기 빼고 데이터 정제
refine_lines = []
for i,line in enumerate(lines):
    refine_lines.append(re.sub('[^ 0-9가-힣a-zA-Z]', '', line))

data = ""
for line in refine_lines:
    data = data + line
print(len(set(data)),set(data))
```

```python
f = open("./dict/nia_refine.txt", 'w')
for line in refine_lines:
    line = line+"\n"
    f.write(line)
f.close()
```
