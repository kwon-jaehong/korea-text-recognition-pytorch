```python
import dataset
import PIL
import os
import random


f = open("./dict/nia_refine.txt", 'r')
lines = f.readlines()
f.close()
for i,line in enumerate(lines):
    lines[i] = line.replace('\n','')

    
    
    
# 최대 문장 길이
max_val = 25

## 문자열을 이어 붙여서 만들기 - 띄어쓰기 포함
concat_dict = [""]
pointer = 0
for i,word in enumerate(lines):
    max_len = random.randint(0,max_val)
    if len(concat_dict[pointer])+len(word)< max_len:
        concat_dict[pointer] =concat_dict[pointer]+' '+word
    else:
        pointer=pointer+1
        concat_dict.append(word)



```

```python
print(concat_dict[:300])
```

```python
f = open("./dict/nia_refine_concat.txt", 'w')
for i in concat_dict:
    f.write(i+"\n")
f.close()
```
