```python
import os
from collections import Counter

f = open("./dict/nia_refine_concat.txt", 'r')
lines = f.readlines()
f.close()
for i,line in enumerate(lines):
    lines[i] = line.replace('\n','')

    
data = ""
for line in lines:
    data = data + line

temp = dict(Counter(data))


data = ""
for line in list(temp.keys()):
    data = data + line
print(data)

f = open("./ch_range.txt", 'w')
f.write(data)
f.close()
```
